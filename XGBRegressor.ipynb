```python
import pandas as pd
import numpy as np
train=pd.read_csv("train.csv")

test=pd.read_csv("test.csv")
trainSex=train["Sex"]
testSex=test["Sex"]
train=train.drop(["Sex"],axis=1)
test=test.drop(["Sex"],axis=1)
train
```

```python
from sklearn.preprocessing import OrdinalEncoder
Encoder=OrdinalEncoder()

train=pd.concat([train,pd.DataFrame(Encoder.fit_transform(trainSex.values.reshape(-1,1)),columns=["Sex"])],axis=1)
test=pd.concat([test,pd.DataFrame(Encoder.fit_transform(testSex.values.reshape(-1,1)),columns=["Sex"])],axis=1)

```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
scatter_matrix(train,figsize=(20,12))
plt.show()
```

```python
traineng=train
traineng=traineng.drop(["Whole weight.1","Whole weight.2","Shell weight"],axis=1)
testeng=test
testeng=testeng.drop(["Whole weight.1","Whole weight.2","Shell weight"],axis=1)
```

```python
traineng["Weight"]=np.log(train["Whole weight"]+train["Whole weight.1"])+(train["Whole weight.2"]+train["Shell weight"])
testeng["Weight"]=np.log(test["Whole weight"]+test["Whole weight.1"])+(test["Whole weight.2"]+test["Shell weight"])
```

```python
traineng=traineng.drop(["Diameter","Height"],axis=1)
testeng=testeng.drop(["Diameter","Height"],axis=1)
```

```python
traineng=traineng.drop(["id"],axis=1)
testeng=testeng.drop(["id"],axis=1)
id=test["id"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaledtraincols=pd.DataFrame(scaler.fit_transform(traineng[["Whole weight","Length"]]),columns=["Whole weight","Length"])
scaledtestcols=pd.DataFrame(scaler.fit_transform(testeng[["Whole weight","Length"]]),columns=["Whole weight","Length"])
```

```python
traineng[["Whole weight","Length"]]=scaledtraincols[["Whole weight","Length"]]
testeng[["Whole weight","Length"]]=scaledtestcols[["Whole weight","Length"]]
```

```python
from sklearn.model_selection import train_test_split
pretrain,pretest=train_test_split(traineng,test_size=0.2,random_state=42,stratify=traineng["Rings"])
```

```python
pretrain
```

```python
xlabels=pretrain["Rings"]
pretrain.drop("Rings",axis=1)
```

```python
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import GridSearchCV
forest=XGBRegressor(random_state=0)
parameters = {
    'n_estimators': [100, 150, 200, 250,300,350,400,450,500],
    'max_depth': [2,3,4,5,6,7,10,11,12,13,14],
}
clf=GridSearchCV(forest,parameters,cv=10,scoring="neg_root_mean_squared_log_error")
```

```python
pretrain=pretrain.drop("Rings",axis=1)
clf.fit(pretrain,xlabels)
```

```python
ylabels=pretest["Rings"]
```

```python
pretest=pretest.drop("Rings",axis=1)
```

```python

preresults=pd.DataFrame(clf.predict(pretest),columns=["Rings"])
```

```python
preresults
```

```python
from sklearn.metrics import mean_squared_log_error as rmlse
rmlse(preresults,ylabels)
```

```python
results=pd.DataFrame(clf.predict(testeng),columns=["Rings"])
```

```python
finalres=pd.concat([id,results],axis=1)
```
