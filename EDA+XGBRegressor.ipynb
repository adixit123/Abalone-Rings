```python
def getfeaturetypes(df,cat=10,car=10):
    cat_vars=[i for i in df.columns if df[i].dtype in ["object","category","bool"]]
    num_cat_vars=[i for i in df.columns if df[i].dtype in ["int64","float64"] and df[i].nunique()<cat]
    car_vars=[i for i in df.columns if df[i].dtype in ["object","category"] and df[i].nunique()>car]
    num_vars=[i for i in df.columns if df[i].dtype in ["int64","float64"] and df[i].nunique()>cat]
    cat_vars=cat_vars+num_cat_vars
    cat_vars= [i for i in cat_vars if i not in car_vars]
    print("Total number of features",len(df.columns))
    print("Number of Categorical Features",len(cat_vars))
    print("Number of Cardinal Features",len(car_vars))
    print("Number of Numeric Features",len(num_vars))
    return cat_vars,car_vars,num_vars
categoricals,cardinals,numerics=getfeaturetypes(X1)
numerics.remove("Rings")
```

```python
print("Categorical Features: ",categoricals)
print("Cardinal Features: ",cardinals)
print("Numerical Features: ",numerics)
```

```python
def outlier_th(df,featname,q1=0.05,q2=0.95):
    q1=df[featname].quantile(q1)
    q3=df[featname].quantile(q2)
    iqr=q3-q1
    lower_lim=q1-1.5*iqr
    upper_lim=q3+1.5*iqr
    return lower_lim,upper_lim
for i in numerics:
    print(i,":",outlier_th(X1,i))
```

```python
outlier_indices=[]
def check_outliers(df,var):
    lower_lim,upper_lim=outlier_th(df,var)
    if df[(df[var]<lower_lim)|(df[var]>upper_lim)].shape[0]>0:
        print(var,": ",df[(df[var]<lower_lim)|(df[var]>upper_lim)].shape[0])
        outlier_indices.append(df[(df[var]<lower_lim)|(df[var]>upper_lim)].index)
        return True
    else:
        return False
for i in numerics:
    print(i,": ",check_outliers(X1,i))
```

```python
def suppress_outliers(df,var):
    lower_lim,upper_lim=outlier_th(df,var)
    df.loc[(df[var]<lower_lim),var]=lower_lim
    df.loc[(df[var]>upper_lim),var]=upper_lim
for i in numerics:
    if check_outliers(X1,i)==True:
        suppress_outliers(X1,i)
        print("Suppressed Outliers for ",i)
    else:
        continue
        
```

```python
import optuna
```

```python
def objective(trial):
    params={"objective":"reg:squaredlogerror",
            "metric":"rmlse",
            "n_estimators":trial.suggest_int("n_estimators",100,1100,step=50),
            "max_depth":trial.suggest_int("max_depth",5,40),
            "learning_rate":trial.suggest_categorical("learning_rate",
                                                      [0.01,0.02,0.025,0.03,0.035,0.05,0.07,0.1,0.2,0.3]),
            "subsample":trial.suggest_float("subsample",0.6,1.0,step=0.1),
            "colsample_bytree":trial.suggest_float("colsample_bytree",0.6,1.0,step=0.1),
            "colsample_bylevel":trial.suggest_float("colsample_bylevel",0.6,1.0,step=0.1)}
    xgbr1=XGBRegressor(**params,random_state=42)
    xgbr1.fit(X1,xlabels)
    results=xgbr1.predict(Y1)
    rmlse=np.sqrt(mean_squared_log_error(results,ylabels))
    return rmlse
```

```python
study=optuna.create_study(direction="minimize")
study.optimize(objective,n_trials=100)
```

```python
study.best_params
```

```python
xgbr_best=XGBRegressor(**study.best_params,random_state=42)
xgbr_best.fit(X1,xlabels)
```

```python
preresults=pd.DataFrame(xgbr_best.predict(Y1),columns=["Rings"])
preresults
```
